```python

minn = []
maxx = []
for k in range(2):
  for j in range(8):
    store1 = []
    for i in range(N):
      oh11_w = w[i][k][j]/b_w[k][j]
      store1.append(oh11_w)
    minn.append(min(store1))
    maxx.append(max(store1))

minn1 = np.mat([minn[:8],
        minn[8:]])
maxx1 = np.mat([maxx[:8],
        maxx[8:]])
```
