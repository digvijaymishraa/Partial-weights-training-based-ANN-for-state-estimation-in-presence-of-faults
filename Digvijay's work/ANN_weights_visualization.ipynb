##**Loading all the different weights**

loading the weights which are saved after training the model with different datasets generating using different values of parameters

```python
w_in_h1=[];b_in_h1=[];w_h1_h2=[];b_h1_h2=[];w_h2_out=[];b_h2_out=[]

for i in range(100):


  a = torch.load('/content/drive/MyDrive/data sets for google collab/project mtech/weights/2_ANN_new/'+str(i+1)+'/w_in_h1.pt',map_location=torch.device('cpu')).detach().numpy()
  bb = torch.load('/content/drive/MyDrive/data sets for google collab/project mtech/weights/2_ANN_new/'+str(i+1)+'/b_in_h1.pt',map_location=torch.device('cpu')).detach().numpy()
  c = torch.load('/content/drive/MyDrive/data sets for google collab/project mtech/weights/2_ANN_new/'+str(i+1)+'/w_h1_h2.pt',map_location=torch.device('cpu')).detach().numpy()
  d = torch.load('/content/drive/MyDrive/data sets for google collab/project mtech/weights/2_ANN_new/'+str(i+1)+'/b_h1_h2.pt',map_location=torch.device('cpu')).detach().numpy()
  e = torch.load('/content/drive/MyDrive/data sets for google collab/project mtech/weights/2_ANN_new/'+str(i+1)+'/w_h2_out.pt',map_location=torch.device('cpu')).detach().numpy()
  f = torch.load('/content/drive/MyDrive/data sets for google collab/project mtech/weights/2_ANN_new/'+str(i+1)+'/b_h2_out.pt',map_location=torch.device('cpu')).detach().numpy()

  w_in_h1.append(a);b_in_h1.append(bb);w_h1_h2.append(c);b_h1_h2.append(d);w_h2_out.append(e);b_h2_out.append(f)
```

###**loading weights which are of model when trained on data at operating condition**

```python
b_w_in_h1 = torch.load('/content/drive/MyDrive/data sets for google collab/project mtech/weights/ANN_orignal_new _model/w_in_h1.pt',map_location=torch.device('cpu')).detach().numpy()
b_b_in_h1 = torch.load('/content/drive/MyDrive/data sets for google collab/project mtech/weights/ANN_orignal_new _model/b_in_h1.pt',map_location=torch.device('cpu')).detach().numpy()
b_w_h1_h2 = torch.load('/content/drive/MyDrive/data sets for google collab/project mtech/weights/ANN_orignal_new _model/w_h1_h2.pt',map_location=torch.device('cpu')).detach().numpy()
b_b_h1_h2 = torch.load('/content/drive/MyDrive/data sets for google collab/project mtech/weights/ANN_orignal_new _model/b_h1_h2.pt',map_location=torch.device('cpu')).detach().numpy()
b_w_h2_out = torch.load('/content/drive/MyDrive/data sets for google collab/project mtech/weights/ANN_orignal_new _model/w_h2_out.pt',map_location=torch.device('cpu')).detach().numpy()
b_b_h2_out = torch.load('/content/drive/MyDrive/data sets for google collab/project mtech/weights/ANN_orignal_new _model/b_h2_out.pt',map_location=torch.device('cpu')).detach().numpy()
```

```python
N = len(w_in_h1)

datasets = np.linspace(117,143,100)
```

#**WEIGHTS BEETWEEN INPUT LAYER AND 1ST HIDDEN LAYER**

```python
for j in range(4):
  plt.figure(figsize=(15,6))
  for i in range(N):
  
    plt.plot(w_in_h1[i][:,j].reshape(-1,1))#,label=str(datasets[i]))
    #plt.plot(w_1[0,:].reshape(-1,1),'k+--',label='129.5*10^6 data')

    plt.xticks(fontsize=20, fontweight='bold')
    plt.yticks(fontsize=20, fontweight='bold')
  
    plt.title('Weights between '+str(j+1)+' th input neuron and neurons of 1st hidden layer',fontsize=20, fontweight='bold')
    plt.xlabel('Weights',fontsize=20, fontweight='bold')
    plt.ylabel('Values of weights',fontsize=20, fontweight='bold')
  
  plt.plot(b_w_in_h1[:,j].reshape(-1,1),'k+--',label='weights for orignal dataset')
  plt.grid()
  #plt.legend(bbox_to_anchor=(1, 1.02),loc='best',fontsize=17)
  lgd = plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left',fontsize = 12)
  plt.autoscale(enable=True, axis='x')
  plt.savefig('True vs filtered: Concentration vs Time instant.png',dpi=200,bbox_extra_artists=(lgd,), bbox_inches='tight')
  # there are 4 column and 8 rows so every column represent values of weight for perticular input neuron
  # so i have plotted column wise vector
```

#**BIASES BETWEEN INPUT AND 1ST HIDDEN LAYER**

```python
plt.figure(figsize=(15,6))
for i in range(N):
  
  plt.plot(b_in_h1[i].reshape(-1,1))
  #plt.plot(w_1[0,:].reshape(-1,1),'k+--',label='129.5*10^6 data')
  
  plt.xticks(fontsize=20, fontweight='bold')
  plt.yticks(fontsize=20, fontweight='bold')
  
  plt.title(' Biases between input layer and 1st hidden layer',fontsize=20, fontweight='bold')
  plt.xlabel('Number Biases',fontsize=20, fontweight='bold')
  plt.ylabel('Values of Biases',fontsize=20, fontweight='bold')
  #plt.legend(bbox_to_anchor=(1, 1.02),loc='best',fontsize=17)
plt.plot(b_b_in_h1.reshape(-1,1),'k+--',label='Bias values for orignal data')
plt.grid()
lgd = plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left',fontsize = 12)
plt.autoscale(enable=True, axis='x')
plt.savefig('True vs filtered: Concentration vs Time instant.png',dpi=200,bbox_extra_artists=(lgd,), bbox_inches='tight')
```
