###**loading weights which are of model when trained on data at operating condition**

```python
b_w_in_h1 = torch.load('/content/drive/MyDrive/data sets for google collab/project mtech/weights/ANN_orignal_new _model/w_in_h1.pt',map_location=torch.device('cpu')).detach().numpy()
b_b_in_h1 = torch.load('/content/drive/MyDrive/data sets for google collab/project mtech/weights/ANN_orignal_new _model/b_in_h1.pt',map_location=torch.device('cpu')).detach().numpy()
b_w_h1_h2 = torch.load('/content/drive/MyDrive/data sets for google collab/project mtech/weights/ANN_orignal_new _model/w_h1_h2.pt',map_location=torch.device('cpu')).detach().numpy()
b_b_h1_h2 = torch.load('/content/drive/MyDrive/data sets for google collab/project mtech/weights/ANN_orignal_new _model/b_h1_h2.pt',map_location=torch.device('cpu')).detach().numpy()
b_w_h2_out = torch.load('/content/drive/MyDrive/data sets for google collab/project mtech/weights/ANN_orignal_new _model/w_h2_out.pt',map_location=torch.device('cpu')).detach().numpy()
b_b_h2_out = torch.load('/content/drive/MyDrive/data sets for google collab/project mtech/weights/ANN_orignal_new _model/b_h2_out.pt',map_location=torch.device('cpu')).detach().numpy()
```

```python
N = len(w_in_h1)
```

#**INPUT and 1st HIDDEN LAYER**

```python
minn = []
maxx = []
for k in range(8):

  for j in range(4):

    store1 = []
    for i in range(N):

      oh11_w = w_in_h1[i][k][j]/b_w_in_h1[k][j]
      store1.append(oh11_w)
    minn.append(min(store1))
    maxx.append(max(store1))

minn2 = np.mat([minn[:4],minn[4:8],minn[8:12],minn[12:16],minn[16:20],minn[20:24],minn[24:28],minn[28:]])
maxx2 = np.mat([maxx[:4],maxx[4:8],maxx[8:12],maxx[12:16],maxx[16:20],maxx[20:24],maxx[24:28],maxx[28:]])
```

```python
in_h1_min = pd.DataFrame(minn2.T,columns=['W1','W2','W3','W4','W5','W6','W7','W8'],index=['input_1','input_2','input_3','input_4'])
in_h1_max = pd.DataFrame(maxx2.T,columns=['W1','W2','W3','W4','W5','W6','W7','W8'],index=['input_1','input_2','input_3','input_4'])

display(in_h1_min)
display(in_h1_max)
```

```python
category1 = []
category2 = []
category3 = []
for j in range(minn2.T.shape[0]):
  for i in range(minn2.T.shape[1]):
  
    if 0.9<=minn2.T[j,i]<=1.1 and 0.9<=maxx2.T[j,i]<=1.1 :
      category1.append('input HL1 W'+str(j+1)+str(i+1))

    elif (0.75<=minn2.T[j,i]<=1.25 and 0.75<=maxx2.T[j,i]<=1.25):
      category2.append('input HL1 W'+str(j+1)+str(i+1))

    else:
      category3.append('input HL1 W'+str(j+1)+str(i+1))
```
